#                                    CASSAVA DATASET

###            The dataset contains information on data points submitted by users around the country describing                                                       observations made in their gardens. 

### Meanings of the column names

##### 1. ID :    Unique identifier for each record.
##### 2. META_INSTANCE_ID:          Unique identifier for each record. Same as ID. (I deleted this column).
##### 3. CREATOR_URI_USER:      Identifier for user who submitted the data
##### 4. CREATION_DATE:    Time when a data record is first recorded in the phone application. Time is device based.
##### 5. LAST_UPDATE_DATE:   Time when a data record is modified. If the record is not modified, this is the same as the 3 (_CREATION_DATE). Time is device based.
##### 6. MARKED_AS_COMPLETE_DATE:   Time when the user saves a data collection record on their device. This is device based.
##### 7. SUBMISSION_TIME:   Time when a user submits a data record to the server. This is device based.
##### 8. START_TIME:   Time when a user starts filling out a data collection form. Time is device based.
##### 9. END_TIME:    The time when the user finalizes a data collection form. Time is device based.
##### 10. SUBMISSION_DATE:  Time when a data record is received on the server. This is server based
##### 11. LAST_UPDATE_URI_USER:  Identifier for a user who updated a record. Only recorded when a record is updated.
##### 12. MODEL_VERSION:  Blank
##### 13. UI_VERSION: Blank
##### 14. IS_COMPLETE: Specifies if the data is received fully. It takes on one of two values 1 and 0. If a data record is incomplete (e.g the image data is invalid) it will record a 0.
##### 15. IMAGE_CLASS: Category describing the type of image captured by the user
##### 16. COMMENT: Extra description that may be provided by the user.
##### 17. DEVICE_ID: Unique identifier for a device.
##### 18. GPS_ALT: GPS Altitude
##### 19. GPS_LAT: GPS Latitude
##### 20. GPS_LNG: GPS Longitude
##### 21. GPS_ACC: GPS Accuracy
##### 22. LINK: Link to the image related to a data record.
